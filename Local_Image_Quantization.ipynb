# Uploading Image

```python
img = cv2.imread("Image-Quantization/Images/Sample 1.png",cv2.COLOR_BGR2RGB)
```

```python
from IPython.display import Image
Image('Image-Quantization/Images/Sample 1.png')
```

# Seperating Colours as one dimensional vectors

```python
r=img[:,:,0]
g=img[:,:,1]
b=img[:,:,2]
r=r.flatten()
g=g.flatten()
b=b.flatten()
img_data=np.dstack((r,g,b))[0]
```

# Importing Kmeans

```python
from sklearn.cluster import MiniBatchKMeans
```

# Function for Applying KMeans to the Pixel Data and reconstructing the Image

```python
def reconstruct(img,bits):
  n=2**bits
  km=MiniBatchKMeans(n,random_state=1)
  labels=km.fit_predict(img_data)
  ri=np.zeros(img.shape)
  for i in range(img.shape[0]):
    t=km.predict(img[i])
    #ri[i,:] = [km.cluster_centers_[l] for l in t]
  return(ri)
```

## Reconstructed Image (4 bit Quantization)

```python
ri4=reconstruct(img,4)
cv2.imwrite('Image-Quantization/Images/4bit Reconstructed1.png',ri4)
Image('Image-Quantization/Images/4bit Reconstructed1.png')
```

## Reconstructed Image (8 bit Quantization)

```python
ri8=reconstruct(img,8)
cv2.imwrite('Image-Quantization/Images/8bit Reconstructed1.png',ri8)
Image('Image-Quantization/Images/8bit Reconstructed1.png')
```

## Reconstructed Image (12 bit Quantization)
